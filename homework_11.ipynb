```python
# Задание 1

# Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
# - оценка 2 и меньше - низкий рейтинг
# - оценка 4 и меньше - средний рейтинг
# - оценка 4.5 и 5 - высокий рейтинг

import pandas as pd
```

```python
data = pd.read_csv('ratings.csv')
data.head()
```

```python
# посчитаем средний рейтинг для каждого Id фильма
data_1 = data.groupby('movieId').mean()[['rating']].sort_values('movieId', ascending=True).head()
data_1.head()
```

```python
def class_movie(rating):
    if rating <= 2:
        return 'Низкий рейтинг'
    elif 2 < rating <= 4:
        return 'Средний рейтинг'
    elif 4.5 <= rating <= 5:
        return 'Высокий рейтинг'        
        
```

```python
data_1['class'] = data_1['rating'].apply(class_movie)
data_1.head()
```

```python
#задание 2


dataset = pd.read_csv('keywords.csv')
```

```python
dataset.loc[120:130]
```

```python
geo_data = {

'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
    
}
```

```python
def add_region(keyword):
    for key, value in geo_data.items():
        for val in value:
            if val in keyword.lower():
                return key
        return 'undefined'
```

```python
dataset['region'] = dataset['keyword'].apply(add_region)
dataset.head()
```
